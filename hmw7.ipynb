```python
import pandas as pd
from surprise import SVD, SVDpp, NMF
from surprise import Dataset
from surprise.model_selection import cross_validate
```

```python
data = Dataset.load_builtin('ml-100k')
```

```python
alg_svd = SVD()
result = cross_validate(alg_svd, data, measures=['RMSE', 'MAE','MSE','FCP'], cv=5, verbose=True)
SVD_result = pd.DataFrame.from_dict(result).mean(axis=0)
```

```python
alg_svdpp = SVDpp()
result_svdpp = cross_validate(alg_svdpp, data, measures=['RMSE', 'MAE','MSE','FCP'], cv=5, verbose=True)
SVDPP_result = pd.DataFrame.from_dict(result_svdpp).mean(axis=0)
```

```python
alg_nmf = NMF()
result_nmf = cross_validate(alg_nmf, data, measures=['RMSE', 'MAE','MSE','FCP'], cv=5, verbose=True)
NMF_result = pd.DataFrame.from_dict(result_nmf).mean(axis=0)
```

```python
Final_result = pd.DataFrame(columns=['SVD', 'SVDpp', 'NMF'])

Final_result['SVD'] = SVD_result
Final_result['SVDpp'] = SVDPP_result
Final_result['NMF'] = NMF_result

Final_result
```
